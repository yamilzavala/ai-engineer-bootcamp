## Create a Chat Model

- Pick a model you have access to (for many accounts, `gpt-4` is a reliable default).
- `temperature=0` makes the output more deterministic.
- `model_kwargs={"seed": 365}` requests deterministic sampling when supported.


```python
from langchain_openai import ChatOpenAI

chat = ChatOpenAI(
    model="gpt-4o-mini",      # change if you prefer another OpenAI chat model
    temperature=0,
    seed=365,
)
```

## Invoke the Model

`invoke()` returns an `AIMessage`. The generated text is available as `response.content`.


```python
prompt = "I've recently adopted a dog. Could you suggest some dog names?"

response = chat.invoke(prompt)
response
```

```python
print(response.content)
```
