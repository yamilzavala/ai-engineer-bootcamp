# Indexing: Document Loading with PyPDF Loader (LangChain 1.2.x)

## Upgrade Notes

- In LangChain 1.2.x, `PyPDFLoader` is provided by **`langchain-community`** (not the core `langchain` package).
- If `from langchain_community.document_loaders import PyPDFLoader` fails, it’s usually due to a **missing dependency** (most commonly `pypdf`) or `langchain-community` not being installed in the active environment.

```python
# Version check (works reliably inside notebooks)
from importlib.metadata import version, PackageNotFoundError

for pkg in ["langchain", "langchain-core", "langchain-community", "langchain-openai", "pypdf"]:
    try:
        print(f"{pkg}: {version(pkg)}")
    except PackageNotFoundError:
        print(f"{pkg}: NOT INSTALLED")
```

## Import The Loader

If the import fails, install the missing packages in the same environment as your notebook kernel, for example:

- `pip install langchain-community pypdf`

```python
from langchain_community.document_loaders import PyPDFLoader
import copy
```

## Load A PDF

Place `Introduction_to_Data_and_Data_Science.pdf` in the same folder as this notebook, or provide the correct path.


```python
loader_pdf = PyPDFLoader("Introduction_to_Data_and_Data_Science.pdf")
```

```python
pages_pdf = loader_pdf.load()
```

```python
pages_pdf
```

```python
pages_pdf_cut = copy.deepcopy(pages_pdf)
```

```python
' '.join(pages_pdf_cut[0].page_content.split())
```

```python
for doc in pages_pdf_cut:
    doc.page_content = " ".join(doc.page_content.split())
```
