## Setting up the API

````python
from openai import OpenAI
import config
````

````python
client = OpenAI(api_key=config.api_key)
````

## Generating Text

1. Modern OpenAI models use the Responses API (not the legacy Completions API).
2. The minimum allowed max_output_tokens is 16.

````python
def generate_text(prompt):
    response = client.responses.create(
        model="gpt-4.1-mini",  # modern, fast, cost-effective
        input=prompt,
        max_output_tokens=16,
        temperature=0.7,
    )
    return response.output_text.strip()
````

````python
prompt = "Once upon a time"
````

````python
generated_text = generate_text(prompt)
print(prompt, generated_text)
````

## Customizing the Output

````python
def generate_text(prompt, max_tokens, temperature):
    response = client.responses.create(
        model="gpt-4.1-mini",
        input=prompt,
        max_output_tokens=max(max_tokens, 16),  # Responses API requires >= 16
        temperature=temperature,
    )
    return response.output_text.strip()
````

````python
generated_text = generate_text(prompt, 50, 0)
print(prompt, generated_text)
````

````python
generated_text = generate_text(prompt, 50, 1)
print(prompt, generated_text)
````

## Poetic Chatbot

````python
def poetic_chatbot(prompt):
    response = client.responses.create(
        model="gpt-4.1-mini",
        input = [
            {
                "role": "system",
                "content": "You are a poetic chatbot."
            },
            {
                "role": "user",
                "content": "When was Google founded?"
            },
            {
                "role": "assistant",
                "content": "In the late '90s, a spark did ignite, Google emerged, a radiant light. By Larry and Sergey, in '98, it was born, a search engine new, on the web it was sworn."
            },
            {
                "role": "user",
                "content": "Which country has the youngest president?"
            },
            {
                "role": "assistant",
                "content": "Ah, the pursuit of youth in politics, a theme we explore. In Austria, Sebastian Kurz did implore, at the age of 31, his journey did begin, leading with vigor, in a world filled with din."
            },
            {
                "role": "user",
                "content": prompt
            }
        ],
        temperature = 1,
        
    )
    return response.output_text.strip()
````

````python
prompt = "When was cheese first made?"
poetic_chatbot(prompt)
````

````python
prompt = "What is the next course to be uploaded to 365DataScience?"
poetic_chatbot(prompt)
````

# Langchain


### 🔄 Updating LangChain: From Legacy Chains & Memory to the Modern Pattern
- The goal is still the same (load data → embed → retrieve → answer)
- The way we wire things together is now more explicit and future-proof

### What is Gone (Legacy LangChain) 🗑️
The following concepts and imports **no longer exist or are deprecated** in modern LangChain:

#### 1. ❌ Chains (`langchain.chains`)
```python
from langchain.chains import ConversationalRetrievalChain
```
- Removed from modern LangChain
- We have no hidden orchestration

#### 2. ❌ Built-in Memory (`ConversationBufferMemory`)
```python
from langchain.memory import ConversationBufferMemory
```
- Removed from modern LangChain
- We have no implicit chat history
- Memory is now explicit, meaning controlled by you

#### 3. ❌ Old OpenAI & Embeddings Imports
```python
from langchain.embeddings import OpenAIEmbeddings
from langchain.llms import OpenAI
from langchain.chat_models import ChatOpenAI
```
- It is moved out of `langchain`
- Now it lives in `langchain-openai`


````python
from langchain_community.document_loaders import WebBaseLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain_community.vectorstores import FAISS
````

<br>

> ⚠ We use a new link that navigates to the 365 Data Science courses page. We’ll use this page to provide context on all available courses on the platform.

````python
url = "https://365datascience.com/courses/"
````

````python
loader = WebBaseLoader(url)
````

````python
raw_documents = loader.load()
````

````python
text_splitter = RecursiveCharacterTextSplitter()
documents = text_splitter.split_documents(raw_documents)
````

````python
embeddings = OpenAIEmbeddings(openai_api_key = config.api_key)
````

````python
vectorstore = FAISS.from_documents(documents, embeddings)
````

````python
retriever = vectorstore.as_retriever()
````

````python
llm = ChatOpenAI(
    openai_api_key=config.api_key,
    model="gpt-4.1-mini",
    temperature=0
)
````

---
#### 💬 How Do We Handle Memory Now?
Instead of:
```python
ConversationBufferMemory(...)
```

We handle the memory ourselves:
```python
chat_history = []
chat_history.append((query, response.content))
```

If needed, you simply inject chat history into the prompt.
This gives you:
- Full control
- Predictable behavior
- No breaking changes

````python
chat_history = []
````

````python
query = "Which course on 365DataScience can help me learn AI?"
````

````python
relevant_docs = retriever.invoke(query)
````

````python
context = "\n\n".join(doc.page_content for doc in relevant_docs)
````

````python
history_text = "\n".join(
    f"User: {q}\nAssistant: {a}" for q, a in chat_history
)
````

````python
prompt = f"""
Use the context below to answer the question.

Conversation history:
{history_text}

Context:
{context}

Question:
{query}
"""

response = llm.invoke(prompt)
````

````python
response = llm.invoke(prompt)
chat_history.append((query, response.content))
````
